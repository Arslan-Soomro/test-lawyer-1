```python
from dsrag.embedding import VoyageAIEmbedding
from dsrag.knowledge_base import KnowledgeBase
from dsrag.llm import AnthropicChatAPI
from dsrag.reranker import VoyageReranker

# Define and initilaize models data for dsRag
file_path = "docs/the-judicial-system-of-pakistan.txt"
kb_id = "judicial-system-of-pakistan"

embedding_model = VoyageAIEmbedding(model="voyage-large-2")
llm = AnthropicChatAPI(model="claude-3-haiku-20240307")
reranker = VoyageReranker(model="rerank-1")


with open(file_path, "r") as f:
    text = f.read()
```

```python
# Create the knowledge base
kb = KnowledgeBase(kb_id=kb_id, embedding_model=embedding_model, auto_context_model=llm, reranker=reranker)
kb.add_document(doc_id=file_path, text=text);
```

```python
search_queries = ["Describe the hierarchy and organizational structure of the judicial system in Pakistan",
                  "Explain the reasons behind the discrepancy in the number of judges across different High Courts in Pakistan",
                  "What are the challenges faced by subordinate courts in Pakistan",
                  "Discuss the types of special courts and administrative tribunals in Pakistan and their jurisdictions",
                  "What are the procedural laws governing civil and criminal proceedings in Pakistan?",
                  "How does the appointment and independence of service tribunal members in Pakistan work?",
                  "Analyze the impact of the 2007 Provisional Constitution Order (PCO) on the strength of judges in the Supreme Court and High Courts of Pakistan",
                  "What is the role of the Federal Shariat Court in Pakistan's judicial system?",
                  "Summarize the annual performance of the judiciary in terms of case disposal and pendency as of 2013",
                  "What is the significance of the Qanun-e-Shahadat Order 1984 in Pakistan’s legal proceedings?"]
results = kb.query(search_queries)
for segment in results:
    print(segment)
```

```python
print(f"Number of Chunks: {len(results)}")
for i, segment in enumerate(results):
    print("----------")
    print(f"Chunk {i+1}:")
    print("----------")
    print(f"Id: {segment['doc_id']}, score: {segment['score']}, chunk_start: {segment['chunk_start']}, chunk_end: {segment['chunk_end']}")
    print(text)
```
